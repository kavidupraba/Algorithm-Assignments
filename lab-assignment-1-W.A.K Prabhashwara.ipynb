### Partitioning an array

We start by showing Python code to reorder the elements of an array so that three elements get sorted: the first element, the element in the middle position and the last element. 

In order to sort the three elements we need to identify the middle position and we need to be able to swap elements. 


````python
def swap(a,i,j):
    a[i],a[j] = a[j],a[i]
    
def sort_three(a):
    n = len(a) - 1
    mid = n // 2
    
    if a[mid] < a[0]:
        swap(a, 0, mid)
    if a[n] < a[0]:
        swap(a, 0, n)
    if a[n] < a[mid]:
        swap(a, mid, n)
````

Now that you know how to sort the three elements, you will see code that rearranges the elements of a slice in an array so that all the elements *smaller or equal* to the middle one (called the pivot) appear before the pivot and all elements *larger or equal* to the pivot appear after the pivot. This is often called partition.

The function ```partition``` re-arranges the array **and** returns the position of the pivot in the array.

````python
# Before start we know that a[lo] <= a[mid] <= a[hi]

def partition(a, lo, hi):
    
    # Place the pivot out of the way, in position hi - 1 
    # The element in position hi is >= pivot so we keep it there! 
    mid = (lo + hi)//2
    swap(a, mid, hi - 1)
    
    # initialize the indices to walk through the array from low to high and from hight to low
    # i starts from the left
    i = lo 
    
    # j starts from the right
    j = hi - 1 
    
    pivot = a[j]
    
    # the algorithm that Tim explains
    while True:
        while True:
            i += 1
            if a[i] >= pivot: break
        # here you know that a[i] >= pivot!
           
        while True:
            j -= 1
            if a[j] <= pivot: break
        # here you know that a[i] >= pivot and a[j] <= pivot
            
        if i >= j: break # you are done if the indices have crossed each other!
            
        # here you know that a[i] >= pivot and a[j] <= pivot and i < j!
        swap(a, i, j)

    # put the pivot in its place and return the position of the pivot    
    swap(a, i, hi - 1)
    return i
````

Finally, here is a program that implements the algorithm. I hope you recognize the programs we used in the exercises 1, 2 and 3. 
    
In Python we usually give names starting with underscore to auxiliary functions.

````python
def quicksort(a):
    _qsort(a,0,len(a)-1)
    
def _qsort(a, lo, hi):
    # For small arrays insertion sort is to be prefered: no recursion.
    # The implementation of sorted in Python uses insertion sort 
    # for arrays of length less than 64. So we can use that. 
    # The base case is then
    #     if lo + 64 > hi: a[lo:hi+1] = sorted(a[lo:hi+1]) 

    # but, at least three elements for the base case 
    # (think of partition, needs 3 elements to work properly)
    
    if lo + 64 > hi: a[lo:hi+1] = sorted(a[lo:hi+1]) 
    else: 
        _pivot(a, lo, hi)
        i = _partition(a, lo, hi)
        _qsort(a, lo, i - 1)
        _qsort(a, i + 1, hi)
        
    
# Now the functions we studied in the previous exercises.
# Here with names as auxiliary functions (starting with _) 
# Also, we use the name _pivot instead of sort_three because it could be 
# some other function than median of three!
# For example, you could pick a random value in the slice as pivot!

def _swap(a,i,j):
    a[i], a[j] = a[j], a[i]
    
def _pivot(a, lo, hi):
    mid = (lo + hi) // 2
    # sort lo, mid, hi:
    if a[mid] < a[lo]:
        _swap(a, lo, mid)
    if a[hi] < a[lo]:
        _swap(a, lo, hi)
    if a[hi] < a[mid]:
        _swap(a, mid, hi)

def _partition(a, lo, hi):
    # place the pivot out of the way, in position hi -1 
    mid = (lo + hi)//2
    _swap(a, mid, hi - 1)
    
    # initialize the indices to walk through the array from low to high and from hight to low
    i = lo
    j = hi - 1
    pivot = a[j]
    
    # the algorithm that Tim explains
    while True:
        while True:
            i += 1
            if a[i] >= pivot: break
           
        while True:
            j -= 1
            if a[j] <= pivot: break
            
        if i >= j: break
        swap(a, i, j)

    # put the pivot in place and return the position of the pivot    
    _swap(a, i, hi - 1)
    return i
````

````python
import math
import random
import time
import pandas as pd
import matplotlib.pyplot as plt


def swap(a,i,j):
    a[i],a[j]=a[j],a[i]

def pviot(a,lo,hi):
    mid=(lo+hi)//2
    if a[mid]<a[lo]:
        swap(a,mid,lo)
    if a[hi]<a[lo]:
        swap(a,mid,lo)
    if a[hi]<a[mid]:
        swap(a,hi,mid)

def partition(a,lo,hi):
    mid=(lo+hi)//2
    swap(a,mid,hi-1)
    i=lo
    j=hi-1
    pviot=a[hi-1]
    while i<j:
        while i<j and a[i]<pviot:
            i+=1
        while i<j and a[j]>pviot:
            j-=1
        if i<j:
            swap(a,i,j)
    swap(a,i,hi-1)
    return i


def quick_sort(a,lo,hi):
    qsort(a,lo,hi-1)

def qsort(a,lo,hi):
    if lo+64>hi:
        a[lo:hi+1]=sorted(a[lo:hi+1])
    else:
        pviot(a,lo,hi)
        i=partition(a,lo,hi)
        qsort(a,lo,i-1)
        qsort(a,i+1,hi)
def tabulate_T_n(start, times):
    table = []
    for t in range(times):
        size = start * (2 ** t)
        total_time = 0
        for i in range(10):
            a = random.sample(range(size * 2), size)
            stime = time.time()
            quick_sort(a, 0, size)
            #print(a)
            #print(size,t,i)
            etime = time.time()
            total_time += etime - stime
        avarage_time = total_time / 10
        table.append((size, avarage_time))
    return table

def main():
    start=800
    times=10
    table=tabulate_T_n(start,times)
    extended_table=[(n,t,math.log2(n),n*math.log2(n),n**2) for n,t in table]
    extended_dic={"size":[],"average_time":[],"BigO(log2n)":[],"Bigo(nlog2n)":[],"Bigo(n**2)":[]}
    for a,b,c,d,e in extended_table:
        extended_dic["size"].append(a)
        extended_dic["average_time"].append(b)
        extended_dic["BigO(log2n)"].append(c)
        extended_dic["Bigo(nlog2n)"].append(d)
        extended_dic["Bigo(n**2)"].append(e)
    table_pd=pd.DataFrame(data=extended_dic)#creating dataframe using pandas

    normalize_table=table_pd.copy()#copying those data frame to new data frame
    # (to normalize data) we do this if deference between scales are too large

    normalize_table["average_time"] /= normalize_table["average_time"].max()
    normalize_table["BigO(log2n)"] /= normalize_table["BigO(log2n)"].max()
    normalize_table["Bigo(nlog2n)"] /= normalize_table["Bigo(nlog2n)"].max()
    normalize_table["Bigo(n**2)"] /= normalize_table["Bigo(n**2)"].max()

    ax=normalize_table.plot(x="size",y=["average_time","BigO(log2n)","Bigo(nlog2n)","Bigo(n**2)"],kind='line',figsize=(10,6))

    ax.set_title("Comparison of average time with BigO complexity ")
    ax.set_xlabel("size")
    ax.set_ylabel("normalize value")
    ax.legend(["average_time","O(log2n)","O(nlog2n)","O(n**2)"])


    #table_pd.plot()
    plt.show()


    print(table_pd)
    file_name="extended_tabel.txt"
    with open(file_name,'w') as f:
        print(table_pd,file=f)


if __name__=="__main__":
    main()

"""
OUT_PUT
     size  average_time  BigO(log2n)  Bigo(nlog2n)    Bigo(n**2)
0     800      0.000200     9.643856  7.715085e+03        640000
1    1600      0.001788    10.643856  1.703017e+04       2560000
2    3200      0.004089    11.643856  3.726034e+04      10240000
3    6400      0.009973    12.643856  8.092068e+04      40960000
4   12800      0.023388    13.643856  1.746414e+05     163840000
5   25600      0.053348    14.643856  3.748827e+05     655360000
6   51200      0.114997    15.643856  8.009654e+05    2621440000
7  102400      0.268185    16.643856  1.704331e+06   10485760000
8  204800      0.721274    17.643856  3.613462e+06   41943040000
9  409600      1.320884    18.643856  7.636523e+06  167772160000

"""


"""
Here used pandas to create charts showing how each math function grows with the size of the list input, and this will help us to determine 
which function closely represents the high-breed quick sort if we use normal charts we can’t  see the differentiation between each function growth so to see 
the difference most up close we will put all the functions output into same range so we can describe and get an idea from data more easily normalization is 
putting a wide range of legends into one range that means between (0-1)  but this might make a visualization effect like legends that have wide range example 
lets take b=[15,25,35,45] after normalization (normalization use this equation (x-x(min))/(x(max)-x(min) but we simplify it by using x/x(max))  b=[0.333,0.555,0.777,1] 
and let's take another set a=[15,20,25,30] after normalizing a=[0.5,0.6,0.83,1] we can see in true values b had higher increment rate of 10 but a only had an increment 
rate of 5 but after normalization “a” have higher increment rate but “b” have a smaller increment rate this happens because normalization compress larger values but stretch 
the smaller values but this will keep the relationship with data points without changing it because of that we can use this method to compare rangers that have more diversity  
so in our hybrid quick sort n**2 will show slower increment rate because of this but log n going to show higher increment rate example( n**2 max value and min value difference 
is large so after division resulting values are much smaller than the previous value but log n  max value and min value difference is small so after division resulting values are 
larger than the previous value ( ex n**2 max=100 n**2 min=10 min/max=10/100=0.1, log n max=50 log n min=10 min/max=10/50=0.2 and 0.2>0.1) this why we can see visualization 
differently but this not effect our decision to make what function the mostly resemble our hybrid quicksort time complexity because we have to find function 
that grow simmiler to T(n) so that function is n log n in this chart we can see average time(T(n))  from a blue line while (n log n)  from a green line 
they have similar growth rate so the function mostly represents the T(n) is “n log2n” – if we have average case scenarios this growth rate mostly align with T(n)
"""
````
